# 地図を描こう
### アクセスされているけど，まだ緯度経度を登録していない地域を検出

```python
df = pd.read_csv("../access_data/ユーザー属性の詳細_地域.csv", encoding="shift-jis")
df = df[df["蝨ｰ蝓"] != "(not set)"]
df = df[df["蟶ょ玄逕ｺ譚"] != "(not set)"]

# 都市名と県名の情報を作成
df["都市名"] = df["蝨ｰ蝓"] + "," + df["蟶ょ玄逕ｺ譚"]
df
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# 1回だけ作る
geolocator = Nominatim(user_agent="myGeocoder", timeout=10)
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

for index, row in df.iterrows():
    location = geocode(row["都市名"])

    if location:
        df.at[index, "緯度"] = location.latitude
        df.at[index, "経度"] = location.longitude
    else:
        df.at[index, "緯度"] = None
        df.at[index, "経度"] = None

    print(f"Processing {index+1}/{len(df)}: {row['都市名']}")
```

```python
df
```

```python
df["freq"] = df['繧｢繧ｯ繝ぅ繝 繝ｦ繝ｼ繧ｶ繝ｼ']

fig = px.scatter_mapbox(
    df,
    lat='緯度',  # 緯度
    lon='経度',  # 経度
    size='freq',
    zoom=4,
    size_max=20,
    color_discrete_sequence=["#00d5ff"]
)

fig.update_layout(
    title_font_size=20,
    geo=dict(
        showcountries=True,  # 国境を表示
        showcoastlines=True,  # 海岸線を表示
        coastlinecolor="Grey",
        landcolor="LightYellow"
    ),
    mapbox_style='carto-darkmatter',
    margin=dict(l=0, r=0, t=0, b=0),
    legend=dict(
        x=1,  # 凡例を右に移動
        y=1,  # 凡例を上に移動
        xanchor="right",  # 凡例の右端を基準に配置
        yanchor="top",  # 凡例の上端を基準に配置
        bgcolor="rgba(255, 255, 255, 0.7)",  # 背景を半透明の白に設定
        bordercolor="Black",  # 枠線の色
        borderwidth=1  # 枠線の幅
    ),
    # mapの初期表示の中心を日本に
    mapbox=dict(
        center=dict(lat=38.5548, lon=137.1529),  # 日本の緯度と経度
    )
)

fig.show()
```

# github風のタイルplot

```python
df = pd.read_csv("../access_data/レポートのスナップショット.csv", encoding="shift-jis")
df["Date"] = pd.to_datetime(df["N 譌･逶ｮ"])
df["log_user"] = df["繧｢繧ｯ繝ゑｿｽｿｽ 繝ｦ繝ｼ繧ｶ繝ｼ"].map(lambda x: np.log(x + 1))
df = df[df["Date"] >= "2024-11-19"]
df
```

```python
fig = calplot(df, x='Date', y="log_user", colorscale='Blues')

# 背景色を黒、見やすいように文字色を白に指定
fig.update_layout(
    paper_bgcolor='#3f3f3f',  # 図全体の周りの背景色
    plot_bgcolor='#3f3f3f',   # プロットエリア（グラフ描画部）の背景色
    font=dict(color='white') # 文字色
)

fig.show()
```
